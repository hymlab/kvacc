### Finding domain offsets of protein seqs for each allele using NetMHCpan's pseudo-sequence
- NetMHCpan에서 작성된 Pseudo sequence를 사용하여 각 MHC alllele의 protein sequence에서의 binding domain의 서열 상의 시작위치를 알아낸다

```python
%run 'mhcseq.ipynb'

def find_offset_for(seq_provider=None, css=None, allele=None, target_seq=None, to=100):
    seq = seq_provider.protein_seq(allele)
    
    for i in range(to):
        poss = sorted(np.unique([cs[1] for cs in css]))
        cur_seq = seq[i:]
        cur_seq = ''.join([cur_seq[j] for j in poss])
        if cur_seq == target_seq:
            return i
    return -1

# seq_provider = MHCAlleleSeqProvider()
# for i in range(netmhcpan_hla_pseq_tab.shape[0]):
#     cur = netmhcpan_hla_pseq_tab.iloc[i]
#     print('Allele: %s, seq: %s' % (cur.allele, cur.seq))
#     offset = find_offset_for(seq_provider=seq_provider, 
#                              css=NETMHCPAN_MHCI_9_CONTACT_SITES, 
#                              allele=cur.allele, 
#                              target_seq=cur.seq)
#     print('Offset: %s' % offset)
```

### Chelvanayagam pMHCI contact sites
- Amino acid sequences for the α1 and α2 domains (residues 1 – 182) for all known HLA-A-encoded class I MHC sequences were obtained from GenBank and aligned tentatively in the absence of insertions and deletions, except for some small stretches of unsequenced residues. The same procedure was performed for all such known HLA-B- and HLA-C-encoded sequences. Polymorphic positions were then identi- fied in all three alignments. Ten different crystal structures of class I MHC peptide nanomer complexes (Table 1) were then examined for the MHC residues in contact with the bound peptide. MHC peptide contacts were defined by selecting residues for which the distance between any atom in the peptide residue and any atom in the MHC residue was less than the sum of their van der Waal’s radii plus a generous distance of <span class="mark">2A ̊</span> (Fig. 1). The additional distance is used to account for the fact that no hydrogen atoms are present in the crystal structures as well as to allow for some variation in sequences that may arise in other potential complexes (vide infra). Contact residues are listed in Table 2. Decamer and octomer complexes were not considered in this study although their inclusion, by treating the decamer P5 residue as a relative insertion and introducing a deletion between P4 and P5 in the octamers, does not drastically alter the results. Contact residues were then scanned to identify the set of MHC residues in the neighborhood of each peptide residue position. This yields a set of nine environments: one for each of the peptide positions P1 to P9. Environments for each peptide position Pχ will be referred to as the Pχ site. For each MHC molecular sequence, the polymorphic compo- nents of these environments are listed in Figures 2, 3, and 4 for HLA-A-, HLA-B-, and HLA-C-encoded sequences, respectively. The positional environments are sorted so that sequences containing iden- tical environments at any peptide position (P1 – P9) are grouped together. Implicit in the grouping of the molecular sequences, at a particular peptide binding site, is that all MHC molecules within a group are predicted to have the same preference of peptide residue at that position. The side chains of polymorphic residues which do not contribute to the antigen binding cleft, such as those projecting down- wards from the cleft floor to form the α3 or β microglobulin interface, are not included.
- {Chelvanayagam:1996ke}, {Luo:2016iw}, {Vang:2017de}

## Using phylogenetically constrained sites as contact sites, such as conserved(or deleterious) sites

- Species별 allele들의 protein sequence를 사용하여 MSA를 생성
- Conserved된 site와 Deleteriousness score(Polyphen score)가 높은 site를 식별하고 sequence polymorphism을 살펴본다 ==> 이 site들이 구조적으로 식별된 contact site일 확률이 유의하게 높은가?(Fisher exact test)
- 높다면 이 site들의 pseudo sequence를 사용하여 학습해보는 것도 좋을 듯...

## PanMHCBindingDomain class

```python
%run 'commons.ipynb'
%run 'mhcinfo.ipynb'
%run 'mhcseq.ipynb'
%run 'aaencoder.ipynb'

import numpy as np
from collections import OrderedDict
from IPython.core.debugger import Tracer

class PanMHCIContactDomain(object):
#     G_ALPHA1_POS = (24, 89)
#     G_ALPHA2_POS = (90, 182)
    DOMAIN_OFFSET_MAP = OrderedDict({
        'HLA-A*02:50': 0,
        'HLA-B*08:03': 0,
        'HLA-A*03:19': 0,
        'HLA-A*24:06': 0,
        'HLA-A*24:13': 0,
        'HLA-B*41:04': 0,
        'HLA-B*44:28': 0,
        'HLA-A': 24, 
        'HLA-B': 24,
        'HLA-C': 24,
        'HLA-E': 21,
        'BoLA-T2c': 21,
        'BoLA-1': 21,
        'BoLA-2': 16,
        'BoLA-3': 8,
        'BoLA-6': 21,
        'H2-K': 21,
        'Mamu-B*017:01': 23,
        'SLA-1': 21,
        'SLA-3': 21,
    })
    DEFAULT_DOMAIN_OFFSET = 24
    
    def __init__(self, mhcseq_provider=MHCAlleleSeqProvider(), add_all_contact_sites=True):
        self.mhcseq_provider = mhcseq_provider
        self._peplen_cs_map = {}
        self._all_hla_sites = []
        
        if add_all_contact_sites:
            self.add_contact_sites(8, IMGT_MHCI_8_CONTACT_SITES)
            self.add_contact_sites(9, IMGT_MHCI_9_CONTACT_SITES)
            self.add_contact_sites(9, NETMHCPAN_MHCI_9_CONTACT_SITES)
            self.add_contact_sites(9, EPICCAPO_MHCI_9_CONTACT_SITES)
            self.add_contact_sites(9, CHELV_MHCI_9_CONTACT_SITES)

    def add_contact_sites(self, pep_len, css):
        if pep_len not in self._peplen_cs_map:
            self._peplen_cs_map[pep_len] = []

        target_cs = self._peplen_cs_map[pep_len]
        for cs in css:
            if cs not in target_cs:
                target_cs.append(cs)

        target_cs.sort(key=lambda cs: cs[0])
        self._update_all_hla_sites()

    def set_contact_sites(self, pep_len, css):
        self._peplen_cs_map[pep_len] = css
        self._update_all_hla_sites()

    def contact_sites(self, pep_len, p_extend=0, h_extend=0):
        old_css = None

        if pep_len in self._peplen_cs_map:
            old_css = self._peplen_cs_map[pep_len]
        else:
            # Consider all AAs of the peptide contact with all HLA contact sites
            old_css = []
            for i in range(pep_len):
                for j in self._all_hla_sites:
                    old_css.append((i, j))
        
        # Extend contact sites
        return self._extend_css(old_css, pep_len, p_extend, h_extend)
    
    def domain_seq(self, allele):
        seq = self.mhcseq_provider.protein_seq(allele)
        offset = self._get_domain_offset(allele)
        return seq[offset:]
        
    def contact_site_seq(self, pep_len, allele, h_extend=0):
        
        seq = self.domain_seq(allele)
        css = self.contact_sites(pep_len, h_extend=h_extend)
        
        sites = sorted(np.unique([cs[1] for cs in css]))
        return ''.join([seq[i] for i in sites])

    def contact_image(self, allele, pep_seq, 
                      p_extend=0, h_extend=0, 
                      p_aaencoder=None, h_aaencoder=None, 
                      width=None, height=None):
        """
        Encode peptide-MHC-II binding data into two-dimensional contact image
        Arguments:
        :param allele: the HLA-II allele
        :param pep_seq: the peptide sequence
        :param h_extend, p_extend: extend contact sites by h_extend(HLA sites) and p_extend(peptide sites)
        :param h_aaencoder: the HLA amino acid encoder
        :param p_aaencoder: the peptide amino acid encoder  
        :param width, height: the width and height of output image
        :return: the two-dimensional image with N channels, where N is the sum of the number of codes of p_aaencoder/h_aaencoder
        """
#         Tracer()()
        pep_len = len(pep_seq)
        h_seq = self.domain_seq(allele)
        css = self.contact_sites(pep_len, p_extend=p_extend, h_extend=h_extend)
        
        h_sites = self._all_hla_sites
        
        ncodes = 0
        if p_aaencoder is not None:
            ncodes += p_aaencoder.ncodes()
        if h_aaencoder is not None:
            ncodes += h_aaencoder.ncodes()
        
        hla_len = len(h_sites)
        mat = np.zeros((ncodes, pep_len, hla_len))
        for cs in css:
            pa = pep_seq[cs[0]]
            ha = h_seq[cs[1]]
            pi = cs[0]
            hi = h_sites.index(cs[1])
            codes = np.asarray([])
            if p_aaencoder is not None:
                codes = np.append(codes, p_aaencoder.encode(pa).values)
            if h_aaencoder is not None:
                codes = np.append(codes, h_aaencoder.encode(ha).values)
            mat[:, pi, hi] = codes 
    
        w_pad = (0, 0)
        h_pad = (0, 0)
        if width is not None:
            w_pad = NumUtils.padsize(width, hla_len)
        if height is not None:
            h_pad = NumUtils.padsize(height, pep_len)
#         print('w_pad:', w_pad)
#         print('h_pad:', h_pad)
        return np.lib.pad(mat, ((0, 0), h_pad, w_pad), 'constant', constant_values=(0, 0)) 

    @property
    def all_hla_sites(self):
        return self._all_hla_sites

#     @property
#     def alleles(self):
#         return self.mhcseq_provider.alleles()
    
    def _update_all_hla_sites(self):
        self._all_hla_sites = []
        for pep_len, css in self._peplen_cs_map.items():
            for cs in css:
                if cs[1] not in self._all_hla_sites:
                    self._all_hla_sites.append(cs[1])
                            
        self._all_hla_sites.sort()

    def _extend_css(self, old_css, pep_len, p_extend=0, h_extend=0):
#         Tracer()()
        css = []
        old_h_sites = sorted(np.unique([cs[1] for cs in old_css]))
        old_h_range = (min(old_h_sites), max(old_h_sites)+1)
        
        for cs in old_css:
            p_begin = p_end = 0
            h_begin = h_end = 0
            if p_extend >= 0:
                p_begin = max(0, cs[0] - p_extend)
                p_end = min(pep_len, cs[0] + p_extend + 1)
            else:
                p_begin = 0
                p_end = pep_len
                
            if h_extend >= 0:
                h_begin = max(0, cs[1] - h_extend)
                h_end = min(old_h_range[1]+100, cs[1] + h_extend + 1)
            else:
                h_begin = old_h_range[0]
                h_end = old_h_range[1]
            
            for cs_i in range(p_begin, p_end):
                for cs_j in range(h_begin, h_end):
                    new_cs = (cs_i, cs_j)
                    if new_cs not in css:
                        css.append(new_cs)
        return css
    
    def _get_domain_offset(self, allele):
        sub_name = MHCAlleleName.parse(allele).format(level=1)
        for k, v in self.DOMAIN_OFFSET_MAP.items():
            if k in allele:
                return v
        return self.DEFAULT_DOMAIN_OFFSET
            
#################################
from unittest import *

class PanMHCIContactDomainTest(TestCase):
    def setUp(self):
        self.cdomain = PanMHCIContactDomain(add_all_contact_sites=False)
        
    def test_add_contact_sites(self):
        expected_css = [(0, 9), (1, 10)]
        self.cdomain.add_contact_sites(9, expected_css)
        
        self.assertTrue(np.array_equal(expected_css, self.cdomain.contact_sites(9)))
        
        expected_css = [(0, 9), (1, 10), (2, 11)]
        self.cdomain.add_contact_sites(9, expected_css)
        self.assertTrue(np.array_equal(expected_css, self.cdomain.contact_sites(9)))

    def test_all_hla_sites(self):
        self.assertTrue(len(self.cdomain.all_hla_sites) == 0)

        self.cdomain.add_contact_sites(9, [(0, 9), (1, 10)])
        
        self.assertTrue(np.array_equal([9, 10], self.cdomain.all_hla_sites))
        
        self.cdomain.add_contact_sites(15, [(0, 7), (1, 9), (2, 12)])
        self.assertTrue(np.array_equal([7, 9, 10, 12], self.cdomain.all_hla_sites))
        
        self.cdomain.set_contact_sites(9, [(0, 11), (1, 8)])
        self.assertTrue(np.array_equal([7, 8, 9, 11, 12], self.cdomain.all_hla_sites))
        
    def test_contact_sites_with_p_extend(self):
        self.cdomain.set_contact_sites(9, [(2, 9)])
        expected_css = [(0, 9), (1, 9), (2, 9), (3, 9), (4, 9)]
        actual_css = self.cdomain.contact_sites(9, p_extend=2)
                             
        self.assertTrue(np.array_equal(expected_css, actual_css))

    def test_contact_sites_with_h_extend(self):
        self.cdomain.set_contact_sites(9, [(2, 9)])
        expected_css = [(2, 7), (2, 8), (2, 9), (2, 10), (2, 11)]

        actual_css = self.cdomain.contact_sites(9, h_extend=2)
        self.assertTrue(np.array_equal(expected_css, actual_css))

    def test_contact_sites_with_no_ref_css(self):
        self.cdomain.set_contact_sites(9, [(2, 9)])
        self.cdomain.set_contact_sites(15, [(0, 8)])
        
        expected_css = []
        pep_len = 12
        for i in range(pep_len):
            expected_css.append((i, 8))
            expected_css.append((i, 9))
            
        actual_css = self.cdomain.contact_sites(pep_len)
                             
        self.assertTrue(np.array_equal(expected_css, actual_css))
    
    def test_contact_image(self):
        allele = 'HLA-A*02:01'
        h_seq = self.cdomain.domain_seq(allele)
        
        self.cdomain.set_contact_sites(9, [(2, 9), (4, 15)])
        self.cdomain.set_contact_sites(10, [(0, 8), (1, 11)])
        
        # Test for 9mer 
        pep_seq = 'WRFLRGYHQ'
        pep_len = len(pep_seq)
#         print('Peptide: %s, %s' % (pep_seq, pep_len))        
        css = self.cdomain.contact_sites(pep_len)
        self.assertTrue(np.array_equal([(2, 9), (4, 15)], css))
        
        h_sites = [8, 9, 11, 15]
        hla_len = len(h_sites)

        sparse_aaencoder = SparseAAEncoder(cmin=0.01, cmax=0.99, dummy=DUMMY_AA)
        sparse_aaencoder.load_code_tab()
        
        cimg = self.cdomain.contact_image(allele=allele, pep_seq=pep_seq, 
                                          p_aaencoder=sparse_aaencoder, h_aaencoder=sparse_aaencoder)
        
        ncodes = sparse_aaencoder.ncodes()*2
        
        expected_cimg_shape = (ncodes, pep_len, hla_len)
        self.assertEquals(expected_cimg_shape, cimg.shape)
                            
        for cs in css:
            pa = pep_seq[cs[0]]
            ha = h_seq[cs[1]]
            pi = cs[0]
            hi = h_sites.index(cs[1])
            expected_codes = np.append(sparse_aaencoder.encode(pa).values, sparse_aaencoder.encode(ha).values)
            self.assertTrue(np.array_equal(expected_codes, cimg[:, pi, hi]))
        
        # Test for 10mer 
        pep_seq = 'WRFLRGYHQR'
        pep_len = len(pep_seq)
#         print('Peptide: %s, %s' % (pep_seq, pep_len))
        css = self.cdomain.contact_sites(pep_len)
        self.assertTrue(np.array_equal([(0, 8), (1, 11)], css))
        
        cimg = self.cdomain.contact_image(allele=allele, pep_seq=pep_seq, 
                                          p_aaencoder=sparse_aaencoder, h_aaencoder=sparse_aaencoder)
                
        expected_cimg_shape = (ncodes, pep_len, hla_len)
        self.assertEquals(expected_cimg_shape, cimg.shape)
        
        for cs in css:
            pa = pep_seq[cs[0]]
            ha = h_seq[cs[1]]
            pi = cs[0]
            hi = h_sites.index(cs[1])
            expected_codes = np.append(sparse_aaencoder.encode(pa).values, sparse_aaencoder.encode(ha).values)
            self.assertTrue(np.array_equal(expected_codes, cimg[:, pi, hi]))
            
    def test_contact_site_seq_equal_to_netmhciipan(self):
        self.cdomain.set_contact_sites(9, NETMHCPAN_MHCI_9_CONTACT_SITES)
        for allele in netmhcpan_hla_pseq_tab.allele:
#             print('Allele: %s' % allele)
            seq1 = None
            try:
                seq1 = self.cdomain.contact_site_seq(9, allele)
            except Exception as e:
                pass
                
            if seq1 is not None:
                seq2 = netmhcpan_hla_pseq_tab.at[allele, 'seq']
#                 print('seq1: %s' % seq1)
#                 print('seq2: %s' % seq2)
                self.assertEquals(seq1, seq2)

##################################
# suite = TestSuite()
# suite.addTests(TestLoader().loadTestsFromTestCase(PanMHCIContactDomainTest))
# TextTestRunner(verbosity=3).run(suite)
```
